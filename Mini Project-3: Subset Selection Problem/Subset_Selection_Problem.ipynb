```python
set = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
subsets1 = []
```

```python
from itertools import combinations
```

```python
counter = 0
for i in combinations(set, 5):
  if (sum(i) == 0):
    subsets1.append(i)
    counter += 1
for j in subsets1:
  print(j)
print("total subsets = ", counter)
```

```python
counter = 0
subsets2 = []

for i in range (3, 7):
  for j in combinations(set, i):
    subsets2.append(j)
    counter += 1

for k in subsets2:
  print(k)
print("total subsets = ",counter)
```
