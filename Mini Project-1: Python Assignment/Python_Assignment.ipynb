```python
import time as t
import random as r
import seaborn as sns

list = []
duration = []
count = []
k = 0

for i in range (5000, 30000, 5000):
  for j in range (0, i):
    list.append(r.random())
  s = t.time()
  list.sort()
  e = t.time()
  duration.append(round(e - s, 10))
  k+=1
  list.clear()
  count.append(i)

print(duration)
print(count)

sns.lineplot(x = count, y = duration)
```

```python
names = {}

students = int(input("Enter number of students: "))

for i in range (0, students):
  name = (input("enter name"))
  marks = []
  for j in range (5):
    m = int(input("enter marks"))
    marks.append(m)
  names[name] = marks

max = -1
min = 101
top = ""
bottom = ""

for k in names:
  total = sum(names[k])
  avg = total / 5
  if(max < avg):
    max = avg
    top = k
  if(min > avg):
    min = avg
    bottom = k

print(top, "is at the top with", max)
print(bottom, "is at the bottom with", min)
```
